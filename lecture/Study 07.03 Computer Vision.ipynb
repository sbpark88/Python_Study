```python
# p.23 이미지 제어 종합 실습
import cv2
def showImage(imgPath):
    # 이미지 불러오기
    img = cv2.imread(imgPath, cv2.IMREAD_COLOR)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(src=gray, dsize=(400, 200))
    cv2.imshow("sparkkorea", gray)

    # 키보드 입력 대시 (0: 무한대기, 이외 밀리세컨드)
    inputKey = cv2.waitKey(0)
    if inputKey == 27: # esc
        cv2.destroyAllWindows()
    elif inputKey == 26: # ctrl + z
        cv2.imwrite("./sparkkorea_out.jpg", gray)
        cv2.destroyAllWindows()
    else:
        cv2.destroyAllWindows()
        print(inputKey)
imgPath = "../exampleCode/images/sparkkorea.jpg"
showImage(imgPath)
```

### 이미지에 Drawing 적용 (사각형)

```python
import cv2
# 이미지 불러오기 및 크기 조절
sampleImg = cv2.imread("../exampleCode/images/sparkkorea.jpg")
resizedImg = cv2.resize(src=sampleImg,
                        dsize=(256,256))

# 불러온 이미지 내 사각형 그리기
cv2.rectangle( resizedImg,
              (80, 50),
              (180,150),
              (0,255,255),
              3)

cv2.imshow("rectImage", resizedImg)

inputKey = cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지에 Drawing 적용 (글자)

#### 글자 입력하기: cv2.putText (“이미지변수명”, 입력문자, (글자 좌측상단좌표), 폰트, 크기, 색상채널, 굵기)

```python
import cv2
# 이미지 불러오기 및 크기 조절
sampleImg = cv2.imread("../exampleCode/images/sparkkorea.jpg")
resizedImg = cv2.resize(src=sampleImg, dsize=(256,256))

# 불러온 이미지 내 사각형 그리기
cv2.putText( resizedImg,   
            "spark",
            (80, 80),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0,255,255),
            3)

cv2.imshow("textImage", resizedImg)
inputKey = cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 영상 불러오기

#### 불러오기: cv2.VideoCapture(파일경로 or 웹캠번호)
#### ret, frame = cap.read()

```python
import cv2
# cap = cv2.VideoCapture("../exampleCode/images/sample_video.mp4")
cap = cv2.VideoCapture(0)

while (cap.isOpened):
    ret, frame = cap.read()
    if ret:
        cv2.imshow("video", frame)
        
        # 30 fps = 1/30 s/f = 1000/30 ms/f = 33 ms/f
        # 30 fps를 만들려면 하나의 프레임(사진)이 33ms동안 표현되어야 한다.
        inputKey = cv2.waitKey(33)
        if inputKey == 27:
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

### 영상 저장하기

```python
import cv2

cap = cv2.VideoCapture("../exampleCode/images/sample_video.mp4")
width, height = int(cap.get(3)), int(cap.get(4))
# fcc = cv2.VideoWriter_fourcc(*'mp4v')
fcc = cv2.VideoWriter_fourcc(*'H264')    # brew install ffmpeg
# fcc = cv2.VideoWriter_fourcc(*'XVID')
video = cv2.VideoWriter("./output.mp4", fcc, 30, (width,height))
while (cap.isOpened):
    
    ret, frame = cap.read()
    try:
        cv2.imshow("video", frame)
    except:
        break
    # 비디오 저장
    video.write(frame)
    inputKey = cv2.waitKey(33)
    if inputKey == 27: # ESC키 종료
        break     
video.release()
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture("../exampleCode/images/sample_video.mp4")
width, height = int(cap.get(3)), int(cap.get(4))
# fcc = cv2.VideoWriter_fourcc(*'mp4v')
fcc = cv2.VideoWriter_fourcc(*'H264')    # brew install ffmpeg
# fcc = cv2.VideoWriter_fourcc(*'XVID')
video = cv2.VideoWriter("./output.mp4", fcc, 30, (width,height))
while (cap.isOpened):
    
    ret, frame = cap.read()
    if ret:
        cv2.imshow("video", frame)
        
        # 30 fps = 1/30 s/f = 1000/30 ms/f = 33 ms/f
        # 30 fps를 만들려면 하나의 프레임(사진)이 33ms동안 표현되어야 한다.
        inputKey = cv2.waitKey(33)
        if inputKey == 27:
            break
    else:
        break

    # 비디오 저장
    video.write(frame)
    inputKey = cv2.waitKey(33)
    if inputKey == 27: # ESC키 종료
        break     
video.release()
cap.release()
cv2.destroyAllWindows()
```
