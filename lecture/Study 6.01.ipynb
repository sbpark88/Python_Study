```python
selloutData.YEARWEEK.max()
```

```python
selloutDataST0002 = selloutData.loc[selloutData.PRODUCTGROUP=='ST0002']
```

```python
selloutDataST0002.YEARWEEK.max()
```

```python
# 연주차 분리
def yearweekSeparation(inValue, option):
    # inValue = 201402
    inValue = str(inValue)
    
    # 1년, 999년, 10000년에 대응하기 위해서 마이너스 인덱스 사용.
    year = inValue[:-2]
    week = inValue[-2:]
    if option =="year":
        value = year
    else:
        value = week
    return value
```

```python
selloutDataST0002["YEAR"] = selloutDataST0002.YEARWEEK.apply(yearweekSeparation, option = "year")
selloutDataST0002["WEEK"] = selloutDataST0002.YEARWEEK.apply(yearweekSeparation, option = "")
selloutDataST0002
```

```python
selloutDataST0002['QTY']=0
```
