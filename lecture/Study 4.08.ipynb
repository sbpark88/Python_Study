### 5. DB 접속 엔진을 만들어 원격지 PostgreSQL, MariaDB DB를 불러오고 수정하고 저장하는 실습.
#### aws postgresql 서버에 접속 한후 “kopo_product_volume” 자료를 불러와서
#### 컬럼명을 소문자로 변경하고 aws mariadb에 저장하세요 파일명은 kopo_product_volume_홍길동

```python
# 모듈 미설치 된 경우 설치해야 할 것들.
# sqlAlchemy : pip install sqlAlchemy : DB 접속 엔진을 만들어준다.
# psycopg2 : pip install psycopg2 : PostgreSQL을 사용 가능하게 해준다.
# pymysql : pip install pymysql : mySQL, MariaDB를 사용 가능하게 해준다.


import pandas as pd
import psycopg2                          # PostgreSQL을 사용 가능하게 해준다.
import pymysql
from sqlalchemy import create_engine     # DB 접속 엔진을 만들어준다.

# PostgreSQL 데이터베이스 접속 엔진 생성.
aws_postgresql_url = 'postgresql://haiteam:haiteam@www.hadoopkorea.com:5432/postgres'
engine_postgresql = create_engine(aws_postgresql_url)

# MariaDB 데이터베이스 접속 엔진 생성.
aws_mariadb_url = 'mysql+pymysql://kopo:kopo@www.hadoopkorea.com:3306/kopo'
engine_mariadb = create_engine(aws_mariadb_url)

# AWS PostgreSQL DB 테이블을 읽어 판다스를 이용해 파이썬에 Data Frame 변수로 저장하기.
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine_postgresql)
# 불러온 DB 확인.
selloutData

# 컬럼해더 재정의.
selloutData.columns = ['salesid','productgroup','yearweek','volume']
# 컬럼해더 변경 확인.
selloutData.head()

# AWS MariaDB에 데이터 저장하기.
# 파일 명으로 넣을 변수 선언.
resultname='kopo_product_volume_박종환'
# 
selloutData.to_sql(name=resultname, con=engine_mariadb, index = False, if_exists='replace')
```
