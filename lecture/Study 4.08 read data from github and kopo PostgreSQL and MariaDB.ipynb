### 5. 자신의 PC에서 csv 파일을 불러와 4. 에서 만든 엔진을 이용해 local DB 서버에 저장하기.

```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine 

# csv 데이터 로딩 후 컬럼 소문자로 변환
selloutData = pd.read_csv("../exampleCode/dataset/kopo_product_volume.csv")
selloutData.columns = ["regionid","productgroup","yearweek","volume"]

print(selloutData.head())
```

```python
# 데이터베이스 접속 엔진 생성
local_postgresql_url = create_engine('postgresql://postgres:postgres@127.0.0.1:812/postgres')    # localhost의 812 포트를 Parallels에 설치된 PostgreSQL 5432 포트로 포트포워딩.

# 데이터 저장
resultname='kopo_product_volume'
selloutData.to_sql(name=resultname, con=local_postgresql_url, index=False, if_exists='replace')    # name, con은 필수 요소라 생략하고 이렇게 써도 된다. (단, 가독성을 위해 넣자.) (.to_sql(resultname, p_engine, index = False, if_exists='replace'))
```

### 5.2 판다스 성능 향상 도구 d6tstack를 활용해보자.

```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine 
import d6tstack

# DB 커넥션 열기
purl = 'postgresql+psycopg2://postgres:postgres@127.0.0.1:812/postgres' 
engine = create_engine(purl) 

# DB 테이블을 읽어 Data Frame 변수에 저장하기
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine) 

selloutData.head()    #디버깅용.
```

```python
# 컬럼해더 재정의
selloutData.columns = ['regionid','pg','yearweek','volume']

# 데이터 저장
resultname='pgresult'
d6tstack.utils.pd_to_psql(df=selloutData, uri=purl, table_name=resultname, if_exists='replace')
```

### 6. 이제 local에서 외부로 나가보자!
#### DB 접속 엔진을 만들어 원격지 PostgreSQL, MariaDB DB를 불러오고 수정하고 저장하는 실습.
#### aws postgresql 서버에 접속하여 “kopo_product_volume” 데이터를 불러온 후, 컬럼명을 소문자로 변경하고, aws mariadb에 저장하세요. 파일명은 "kopo_product_volume_홍길동".

```python
# 모듈 미설치 된 경우 설치해야 할 것들.
# sqlAlchemy : pip install sqlAlchemy : DB 접속 엔진을 만들어준다.
# psycopg2 : pip install psycopg2 : PostgreSQL을 사용 가능하게 해준다.
# pymysql : pip install pymysql : mySQL, MariaDB를 사용 가능하게 해준다.


import pandas as pd
import psycopg2                          # PostgreSQL을 사용 가능하게 해준다.
import pymysql
from sqlalchemy import create_engine     # DB 접속 엔진을 만들어준다.

# PostgreSQL 데이터베이스 접속 엔진 생성.
aws_postgresql_url = 'postgresql://haiteam:haiteam@www.hadoopkorea.com:5432/postgres'
engine_postgresql = create_engine(aws_postgresql_url)

# MariaDB 데이터베이스 접속 엔진 생성.
aws_mariadb_url = 'mysql+pymysql://kopo:kopo@www.hadoopkorea.com:3306/kopo'
engine_mariadb = create_engine(aws_mariadb_url)

# AWS PostgreSQL DB 테이블을 읽어 판다스를 이용해 파이썬에 Data Frame 변수로 저장하기.
selloutData = pd.read_sql_query('SELECT * FROM kopo_product_volume', engine_postgresql)
# 불러온 DB 확인.
selloutData

# 컬럼해더 재정의.
selloutData.columns = ['salesid','productgroup','yearweek','volume']
# 컬럼해더 변경 확인.
selloutData.head()

# AWS MariaDB에 데이터 저장하기.
# 파일 명으로 넣을 변수 선언.
resultname='kopo_product_volume_홍길동'
# 
selloutData.to_sql(name=resultname, con=engine_mariadb, index = False, if_exists='replace')
```
