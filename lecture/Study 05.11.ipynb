# 1. 리스트로부터 데이터프레임 생성 (복습)

```python
import pandas as pd
```

```python
productList = ['ST0001','ST0002','ST0003']
yearweekList = ['202001','202002','202003']
qtyList = ['10','12','15']
```

```python
pd.DataFrame(zip(productList,yearweekList, qtyList),
             columns=['product','yearweek','qty'],
             index=['one','two','three'])
```

```python
# 방법 1. 여러개의 List를 zip으로 묶어서 데이터프레임 만들기.
pd.DataFrame(zip(productList,yearweekList,qtyList),
            columns=['product','yearweek','qty'])
```

```python
# 방법 2. 딕셔너리로 묶어서 데이터프레임 만들기.
pd.DataFrame({'product':productList,'yearweek':yearweekList,'qty':qtyList})
```

```python
# 방법 3. 각각 데이터프레임 만든 다음 concat으로 합치기.
productDF = pd.DataFrame(productList, columns=['product'])
yearweekDF = pd.DataFrame(yearweekList, columns=['yearweek'])
qtyDF = pd.DataFrame(qtyList, columns=['qty'])
pd.concat([productDF,yearweekDF,qtyDF], axis=1)
```

```python
# 각각의 데이터를 DF로 만들고, pd.concat을 사용하는게 정석이다. 더 다양한 작업을 할 수 있다.
# 단, 인덱스가 다를 경우 초기화 해서 붙여줘야 하기 때문에 귀찮아서 크롤링에서는 zip을 많이 사용한다.
```

```python
productDf = pd.DataFrame(productList, columns=['product'])
yearweekDf = pd.DataFrame(yearweekList, columns=['yearweek'])
qtyDf = pd.DataFrame(qtyList, columns=['qty'])
result = pd.concat([productDf,yearweekDf,qtyDf], axis=1)
result
```

```python

**Getting values on a DataFrame with an index that has integer labels**

Another example using integers for the index

>>> df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
...      index=[7, 8, 9], columns=['max_speed', 'shield'])
>>> df
   max_speed  shield
7          1       2
8          4       5
9          7       8

Slice with integer labels for rows. As mentioned above, note that both
the start and stop of the slice are included.

>>> df.loc[7:9]
   max_speed  shield
7          1       2
8          4       5
9          7       8

**Getting values with a MultiIndex**

A number of examples using a DataFrame with a MultiIndex

>>> tuples = [
...    ('cobra', 'mark i'), ('cobra', 'mark ii'),
...    ('sidewinder', 'mark i'), ('sidewinder', 'mark ii'),
...    ('viper', 'mark ii'), ('viper', 'mark iii')
... ]
>>> index = pd.MultiIndex.from_tuples(tuples)
>>> values = [[12, 2], [0, 4], [10, 20],
...         [1, 4], [7, 1], [16, 36]]
>>> df = pd.DataFrame(values, columns=['max_speed', 'shield'], index=index)
>>> df
                     max_speed  shield
cobra      mark i           12       2
           mark ii           0       4
sidewinder mark i           10      20
           mark ii           1       4
viper      mark ii           7       1
           mark iii         16      36

Single label. Note this returns a DataFrame with a single index.

>>> df.loc['cobra']
         max_speed  shield
mark i          12       2
mark ii          0       4

Single index tuple. Note this returns a Series.

>>> df.loc[('cobra', 'mark ii')]
max_speed    0
shield       4
Name: (cobra, mark ii), dtype: int64

Single label for row and column. Similar to passing in a tuple, this
returns a Series.

>>> df.loc['cobra', 'mark i']
max_speed    12
shield        2
Name: (cobra, mark i), dtype: int64

Single tuple. Note using ``[[]]`` returns a DataFrame.

>>> df.loc[[('cobra', 'mark ii')]]
               max_speed  shield
cobra mark ii          0       4

Single tuple for the index with a single label for the column

>>> df.loc[('cobra', 'mark i'), 'shield']
2

Slice from index tuple to single label

>>> df.loc[('cobra', 'mark i'):'viper']
                     max_speed  shield
cobra      mark i           12       2
           mark ii           0       4
sidewinder mark i           10      20
           mark ii           1       4
viper      mark ii           7       1
           mark iii         16      36

Slice from index tuple to index tuple

>>> df.loc[('cobra', 'mark i'):('viper', 'mark ii')]
                    max_speed  shield
cobra      mark i          12       2
           mark ii          0       4
sidewinder mark i          10      20
           mark ii          1       4
viper      mark ii          7       1
```

## 숫자 & 문자 조건도 검색 되는데 굳이 왜 임시로 바꿀까?
### "201702"라는 데이터가 있다. 일반적으로 문자열 처리를 많이 하기 때문에 기본 컬럼 데이터 타입은 문자열로 두고, 이것을 계산에 사용할 때 yearweek > 201702 이런 계산이 필요할 경우 잠시 int 타입으로 바꿔서 계산한다. 데이터 자체를 바꾸기에는 문자열 데이터를 더 많이 쓰기 때문에 이렇게 사용함.

## 리스트를 통해 데이터프레임 조회하기

```python
emiCondition = ['2','3']
customerData.loc[customerData.EMI.isin(emiCondition)]    # is in : emiCondition이라는 리스트에 의해 EMI가 '2', '3'인 데이터를 조회.
```

```python
customerData.loc[~customerData.EMI.isin(emiCondition)]    # ~ 에 의해 반전된다. EMI가 '2', '3'이 아닌 데이터를 조회.
```
