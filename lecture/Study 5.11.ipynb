# 1. 리스트로부터 데이터프레임 생성 (복습)

```python
import pandas as pd
```

```python
productList = ['ST0001','ST0002','ST0003']
yearweekList = ['202001','202002','202003']
qtyList = ['10','12','15']
```

```python
pd.DataFrame(zip(productList,yearweekList, qtyList),
             columns=['product','yearweek','qty'],
             index=['one','two','three'])
```

```python
# 각각의 데이터를 DF로 만들고, pd.concat을 사용하는게 정석이다. 더 다양한 작업을 할 수 있다.
# 단, 인덱스가 다를 경우 초기화 해서 붙여줘야 하기 때문에 귀찮아서 크롤링에서는 zip을 많이 사용한다.
```

```python
productDf = pd.DataFrame(productList, columns=['product'])
yearweekDf = pd.DataFrame(yearweekList, columns=['yearweek'])
qtyDf = pd.DataFrame(qtyList, columns=['qty'])
result = pd.concat([productDf,yearweekDf,qtyDf], axis=1)
result
```

## 숫자 & 문자 조건도 검색 되는데 굳이 왜 임시로 바꿀까?
### "201702"라는 데이터가 있다. 일반적으로 문자열 처리를 많이 하기 때문에 기본 컬럼 데이터 타입은 문자열로 두고, 이것을 계산에 사용할 때 yearweek > 201702 이런 계산이 필요할 경우 잠시 int 타입으로 바꿔서 계산한다. 데이터 자체를 바꾸기에는 문자열 데이터를 더 많이 쓰기 때문에 이렇게 사용함.

## 리스트를 통해 데이터프레임 조회하기

```python
emiCondition = ['2','3']
customerData.loc[customerData.EMI.isin(emiCondition)]    # is in : emiCondition이라는 리스트에 의해 EMI가 '2', '3'인 데이터를 조회.
```

```python
customerData.loc[~customerData.EMI.isin(emiCondition)]    # ~ 에 의해 반전된다. EMI가 '2', '3'이 아닌 데이터를 조회.
```
